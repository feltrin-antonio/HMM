## The program works if we consider subsets of 100 elements max...

```python
params = np.array([[-1, 0.5], [1., 0.5]])
A = np.array([[0.92, 0.08], [0.03, 0.97]]) # taken by an article
PI = np.array([0.5, 0.5])
X_tot = []
for i in range(0, data.size, 50):
    print(f"A = {A}")
    print(f"Pi = {PI}")
    print(f"params = \n {params}")
    hidden = HMM(2, A, PI, data[i:i+50], params)
    mean, std, X, A, PI = hidden.iteration(show=False)
    params = np.stack((mean, std)).T
    X_tot.extend(X)

X = [mean[0] if x == 0 else mean[1] for x in X_tot]
plt.plot(time, data)
plt.plot(time, X)
```
